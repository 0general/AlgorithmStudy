```python
N = int(input()) # input()값은 문자열이니까 int로 타입 변경
n = []

for i in range(N):       # 입력한 값만큼 for문을 돌림
    a = int(input())     # 이때 들어가는 입력값들도 int 타입으로 받을 수 있게 설정
    n.append(a)          # 빈 list에 추가 시켜서 원하는 값이 들어간 list 생성

b = sorted(n)            # sorted 함수를 써서 정렬

for i in b:              # 각 줄로 출력될 수 있도록 for문 설정
    print(i)
```

```python
#백준 전용
import sys

N = int(sys.stdin.readline())
n = []

for i in range(N):
    a = int(input())
    n.append(a)

b = sorted(n) 

for i in b:
    print(i)
```
