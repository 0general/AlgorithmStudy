```python
N = int(input())
n = []

for i in range(N):
    a = int(input())
    n.append(a)

# 선택 정렬
for i in range(N):
    min_index = i                              # 가장 작은 원소의 인덱스 #range(5)니까 0으로 가정
    for j in range(i+1, N):                    # range(1, 5)~range(5, 5)가 될 때까지 계속 for문 돌림
        if n[min_index] > n[j]:                # 만약 n[0] = 5 가 n[1] = 2 보다 크다면 가장 작은 원소의 자리에 j 값이 들어감
            min_index = j                      # 5 > 2이므로 2가 min_index 자리에 들어감 => 이런 식으로 비교를 반복
    n[i], n[min_index] = n[min_index], n[i]    # 작은 수가 앞에 와야 하니 자리 스와핑

for i in n:
    print(i)
```

```python
#백준 전용
import sys

N = int(sys.stdin.readline())
n = []

for i in N:
    a = int(input())
    n.append(a)

b = sorted(n) 

for i in b:
    print(i)
```
